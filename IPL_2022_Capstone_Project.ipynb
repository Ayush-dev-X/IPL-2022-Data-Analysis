1)Which team won the most matches?

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y = match_wins.index,x = match_wins.values,palette='rainbow')
plt.title("Most matches win by a team")
```

### Observation

Gujarat Titans won the highest number of matches in IPL 2022, demonstrating consistent performances throughout the tournament and eventually becoming the champions.

2)Toss Decision Trends

```python
sns.countplot(x = df['toss_decision'],palette='rainbow')
plt.title("Toss Decisions Trends")
```

### Observation

The toss decision analysis shows that teams generally opted to **field first** rather than bat first. This suggests that captains believed chasing a target offered a strategic advantage under the playing conditions.

3)Toss Winner vs Match Winner

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
peercentage = (count/df.shape[0] * 100)
peercentage.round(2)
```

### Observation

The results indicate that toss victories did not consistently lead to match victories. Overall team performance played a more significant role in determining the outcome.

4)How do teams win? (Runs vs Wickets)

```python
sns.countplot( x = df['won_by'])
plt.title("won by")
```

### Observation

Teams secured victories by both **runs** and **wickets**, although wins by wickets were slightly more frequent. This indicates that while defending totals remained effective, chasing was marginally more successful in IPL 2022.

## Key Player Performances

1)Most "Player of the Match" Awards

```python
count = df['player_of_the_match'].value_counts().head(10)
```

```python
sns.barplot(y = count.index,x = count.values,palette='mako')
plt.title("Player of the match")
```

### Observation

The player with the highest number of Player of the Match awards consistently delivered impactful performances and contributed significantly to their team's success.

2)Top Scores

```python
df.head()
```

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind = 'barh')
```

### Observation

The highest individual score reflects an exceptional batting performance and highlights the player's ability to dominate the opposition.

10 best bowling figures

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers
```

```python
top_bowlers.plot(kind = 'barh')
```

### Observation

The top 10 bowling performances highlight the significant impact bowlers had on match outcomes in IPL 2022. Several bowlers delivered match-winning spells by taking multiple wickets while conceding very few runs, demonstrating the importance of disciplined and effective bowling in the tournament.
